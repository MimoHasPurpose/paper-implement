```python
import numpy as np

def _read_pose_txt(path):
    mat = np.loadtxt(path, dtype=np.float32)  # shape (3,4)
    if mat.shape == (3, 4):
        R, t = mat[:, :3], mat[:, 3]
        return R, t
    return None
```

```python
# e.g., pose0.txt, pose1.txt, ...
# pdb.set_trace()
for f in files:
    if f.startswith("pose") and f.endswith(".txt"):
        num = int(f[4:-4])  # extract the number after "pose" and before ".txt"
        new_name = f"{num:06d}.txt"  # zero-pad to 6 digits
        old_path = os.path.join(pose_dir, f)
        new_path = os.path.join(pose_dir, new_name)
        os.rename(old_path, new_path)
        print(f"{f} -> {new_name}")
```
